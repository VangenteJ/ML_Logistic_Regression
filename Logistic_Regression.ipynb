```python
# Importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating my own data set
df = pd.DataFrame(np.random.rand(20, 9), columns=['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label'])
df
```

```python
# Checking dataset
df.head()
```

```python
# Select row and change column values to 0s and 1s.
features_c = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
x = df[features_c]
y = df[df>.4]
y = y.fillna(value=1)
y = y[y>.99]
y = y.fillna(value=0)
y = y.label
y
```

```python
# Create validation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# Create our model
from sklearn.linear_model import LogisticRegression
lrg = LogisticRegression()
lrg.fit(x_train, y_train)

prediction = lrg.predict(x_test)
```

```python
# Import metrics and create confusion matrix
from sklearn import metrics
c_matrix = metrics.confusion_matrix(y_test, prediction)
c_matrix
```

```python
# Check accuracy
print (metrics.accuracy_score(y_test, prediction))
```
